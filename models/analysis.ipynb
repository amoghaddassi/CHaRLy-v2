```python
#get some sim results
agent = FlatAgentPseudoReward(.2)
r, a = main(10000, agent)
```

```python
def plot_rolling_mean(r):
    #gets the rolling mean of rewards
    df = pd.DataFrame(r)
    mean = df.rolling(50).mean()
    plt.plot(range(len(mean)), mean)
    plt.xlabel("Trial")
    plt.ylabel("Average reward over last 50 trials")
```

```python
plot_rolling_mean(r)
```
